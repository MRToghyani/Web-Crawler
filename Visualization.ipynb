```python
import time
import mysql.connector
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def connect():

    # Connect to the database
    connection = mysql.connector.connect(
        host="host",
        user="user",
        password="password",
        database="database"
    )

    # Create a cursor
    cursor = connection.cursor()

    # Execute a query to retrieve data
    query = "SELECT source, target FROM link"
    cursor.execute(query)

    # Fetch the data
    data = cursor.fetchall()

    # Close the cursor and connection
    cursor.close()
    connection.close()
    
    return data
```

```python
data = connect()
# data
```

```python
G = nx.Graph()
G.add_edges_from(data)
```

```python
# G.edges()
```

```python
length = len(data)
print("All Data : ", length)
print("Nodes number: " , len(G.nodes()))
print("Edges number: " , len(G.edges()))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(50,50))
subax = plt.subplot(111)
options = {"node_color": "black", "node_size": 50, "linewidths": 0, "width": 0.1}

pos = nx.spring_layout(G, seed=1969)  # Seed for reproducible layout
nx.draw(G, pos, **options,ax=subax)   #,ax=subax
plt.savefig("links_graph.png")
plt.show()
```
